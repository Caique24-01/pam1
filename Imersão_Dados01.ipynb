**Desafio 01**

Ordenar o gráfico countplot

```python
sns.set()
plt.figure(figsize=(10, 8))
ax = sns.countplot(x ='droga',data=dados.query('droga in @compostos'), order = compostos)
ax.set_title('Top 5 Drogas')
plt.show()
```

**Desafio 02**

Melhorar a parte visualização do gráfico alterandoo tamanha da fonte

```python
sns.set()
plt.figure(figsize=(10, 8))
ax = sns.countplot(x ='droga',data=dados.query('droga in @compostos'), order = compostos)
ax.set_title('Top 5 Drogas', size = 20 )
ax.set_ylabel('Contagem', size = 16)
ax.set_xlabel('Composto', size = 14)
plt.show()
```

Em suma, a informação deve ser de fácil leitura e interpretação. Um dado oculto, um gráfico com dimensioamento ruim, uma escala desproporcional... todos estes fatores podem afetar negativamente a interpretação, prejudicando a tomada de decisão e análises necessárias.
